напишіть функцію гіпотези лінійної регресії у векторному вигляді

```python
import numpy as np
def hypothesis(X, w):
    return np.dot(X, w)
```

створіть функцію для обчислення функції втрат у векторному вигляді

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

реалізуйте один крок градієнтного спуску

```python
def gradient_descent(X, y, learning_rate=0.01, num_iterations=1):
    m, n = X.shape
    w = np.zeros(n)

    for _ in range(num_iterations):
        predictions = hypothesis(X, w)
        errors = predictions - y

        gradient = 2 / m * np.dot(X.T, errors)
        w -= learning_rate * gradient

        mse = mean_squared_error(y, predictions)
        print(f"Iteration: {_}, Mean Squared Error: {mse}, w={w}")

    return w
```

```python
import pandas as pd

file_path = '/content/Housing.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
pred = df['price']
x = df[['area', 'bathrooms', 'bedrooms']].values

print(gradient_descent(x, pred, num_iterations=1000))
```

```python
x_b = x
w_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(pred)
print(w_best)
```

знайдіть ці ж параметри за допомогою аналітичного рішення

```python
import pandas as pd

file_path = '/content/Housing.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
pred = df['price']
x = df[['area', 'bathrooms', 'bedrooms']].values

print(gradient_descent(x, pred, num_iterations=1000))
```

```python
x_b = np.c_[np.ones((len(x), 1)), x]
w_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(pred)
print(w_best)
```
